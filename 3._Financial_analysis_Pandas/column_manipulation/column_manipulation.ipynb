```python
import pandas as pd
from pathlib import Path
```

# Column Manipulation

```python
csvpath = Path("../Resources/customers.csv")
```

```python
customer_dataframe = pd.read_csv(csvpath)
customer_dataframe.head()
```

### Replace Columns

```python
# Display column names
customer_dataframe.columns
```

```python
# Rewrite the column names
columns = ["Full Name", "E mail", "Address", "Zip Code", 
            "Credit Card Number", "Account Balance"]

customer_dataframe.columns = columns
customer_dataframe.head()
```

### Rename Columns

```python
customer_dataframe=  customer_dataframe.rename(columns={
    "Full Name":"full_name",
    "Credit Card Number":"Credit_card_number"
})  
```

### Reorder Columns

```python
customer_dataframe = customer_dataframe[["full_name", "Credit_card_number", "E mail", "Address", "Zip Code", 
        "Account Balance"]]
customer_dataframe.head()  
```

### Create Columns

```python

customer_dataframe["Balance (1k)"] = customer_dataframe["Account Balance"]/1000
customer_dataframe.head()
```

### Split Columns

```python
customer_dataframe["full_name"].str.split(" ", expand=True)
```

```python
names = customer_dataframe["full_name"].str.split(" ", expand=True)
names.head()
```

```python
customer_dataframe["first_name"] = names[0]
customer_dataframe["last_name"] = names[1]
customer_dataframe.head()
```

### Delete Columns

```python
customer_dataframe =customer_dataframe.drop(columns=["full_name"])
customer_dataframe
```

#### Exporting a Dataframe

```python
customer_dataframe.to_csv("../Resources/customer_reworked.csv")
```
